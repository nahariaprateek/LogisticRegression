```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##Data Collection and Data Processing Step 

```python
#loading dataset to pandas 
sonar_data = pd.read_csv('/content/sonar_data.csv', header = None)
```

```python
sonar_data.head()
```

```python
#number of rows and columns  & description

sonar_data.shape
sonar_data.describe()
```

```python
#how many examples are there for mine and how many are there for rock
sonar_data[60].value_counts()
```

#From the above we can see that -  Rocks are 97 and 111 for mine.


```python
sonar_data.groupby(60).mean()
```

```python
#Seperating data and lebels - dropping the 60th col and only 60th col in X and Y
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]

```

```python
#Now split this data into training and test data 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

Model Training -> Logistic Regression


```python
model = LogisticRegression()
```

Training the logistic regression model with training dataset

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data = ', training_data_accuracy)
```

```python
#accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on testing data = ', testing_data_accuracy)
```
